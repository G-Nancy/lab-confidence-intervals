# Challenge 3 - Constructing Confidence Intervals

We will test whether the hourly wage of all hourly workers is significantly different from $30/hr.

In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. Is $30/hr within that interval?

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here
print('true mean: ',salaries['Hourly Rate'].mean(),'true sigma: ',salaries['Hourly Rate'].std(ddof=0))

def sigma(obs):
    num = np.sum((np.mean(obs)-obs)**2)
    den = len(obs)
    print(den)
    sigma_biased = np.sqrt(num/den)
    sigma_unbiased = np.sqrt(num/(den-1))
    return sigma_biased, sigma_unbiased

print(' sigma biased and unbiased: ', sigma(salaries['Hourly Rate']))
```

```python
# 95.5% confidence needs: Mean, number of observation, std of original
salaries_hourly_notna = salaries[salaries['Hourly Rate'].notna()]

print('true mean: ',salaries_hourly_notna['Hourly Rate'].mean(),'true sigma: ',salaries_hourly_notna['Hourly Rate'].std(ddof=0))
print(' sigma biased and unbiased: ', sigma(salaries_hourly_notna['Hourly Rate']))
```

### Confidence Interval

```python
salaries_hr_sample = salaries_hourly_notna.sample(1000)
salaries_hr_sample_mean = salaries_hr_sample['Hourly Rate'].mean()
salaries_hr_sample_std = salaries_hr_sample['Hourly Rate'].std()
salaries_df = len(salaries_hr_sample['Hourly Rate']) # degress of freedom

# Confidence interval with z-score
confidence_interval = 0.95
z = norm.ppf(confidence_interval) # 0.95 is confidence interval
t = st.t.ppf(1-((1-confidence_interval)/2), salaries_df-1)

print('salaries_hr_sample_mean x:', salaries_hr_sample_mean, '\n salaries_hr_sample_std s:', salaries_hr_sample_std, '\n degrees of freedom df:',salaries_df)

# Calculate 95% confidence interval
error_margin = 2*(salaries_hr_sample_std/np.sqrt(salaries_df))
z_error_margin = z*(salaries_hr_sample_std/np.sqrt(salaries_df))
t_error_margin = t*(salaries_hr_sample_std/np.sqrt(salaries_df))


print('\n 95% Confidence interval is :', salaries_hr_sample_mean-error_margin, "-", salaries_hr_sample_mean+error_margin)

# Using z-score python function 
print('\n 95% Confidence interval with z stats function is :', norm.interval(confidence_interval, loc=salaries_hr_sample_mean, scale=salaries_hr_sample_std/np.sqrt(salaries_df)))

print(f'95% Confidence interval(z-score:{z}) is :', salaries_hr_sample_mean-z_error_margin, "-", salaries_hr_sample_mean+z_error_margin)

# Using t-score python function 
print('\n95% Confidence interval with t stats function is :', st.t.interval(confidence_interval,salaries_df-1, loc=salaries_hr_sample_mean, scale=salaries_hr_sample_std/np.sqrt(salaries_df)))
print(f'95% Confidence interval(t-score:{t}) is :', salaries_hr_sample_mean-t_error_margin, "-", salaries_hr_sample_mean+t_error_margin)
```

### Computing number of samples needed

Confidence interval = [μ - t * s / √n , μ + t * s / √n]

W =   (μ + t * s / √n) - (μ - t * s / √n) = 2 * t * s / √n


Objective: n=?

W = 2 * t * s / √n         ⇔

√n * W = 2 * t * s         ⇔

√n = 2 * t * s / W         ⇔

n = (2 * t * s / W)2


```python
def find_n(t,s,W):
    return (2*t*s/W)**2

def find_t(conf, n):
    return st.t.ppf(1-((1-conf)/2), n-1)

# Confidence interval with z-score
confidence_interval = 0.95
t0 = norm.ppf(1-(1-confidence_interval)/2)
s = salaries_hr_sample['Hourly Rate'].std()
W = 5

n_tmp = find_n(t0,s,W)
t1 = find_t(confidence_interval, n_tmp)
n1_tmp = find_n(t1,s,W)
t2 = find_t(confidence_interval, n1_tmp)
n2_tmp = find_n(t2,s,W)

print(f'beginning values - t0:{t0}, std:{s}, error_margin:{W}')
print(f' test - n0:{n_tmp},n1:{n1_tmp},n1:{n2_tmp} ')
```
